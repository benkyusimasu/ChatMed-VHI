```python
#GPT_zero_prompt
import openai
import json
import time


client = openai.OpenAI(api_key="Your_key")  


with open("test.json", "r", encoding="utf-8") as f:
    data = [json.loads(line) for line in f]
sample_data = data

predictions = {}


for idx, item in enumerate(sample_data):
    prompt = f"""You are an assistant that extracts answers from biomedical texts.

Extract the answer to the following question using **only the words from the given context**.
Answer with the **shortest span** from the context that **directly answers** the question.
Do **not include any additional words or explanation**.

Context:
{item['context']}

Question:
{item['question']}

Answer:"""

    try:
        response = client.chat.completions.create(
            model="gpt-3.5-turbo",
            messages=[
                {"role": "system", "content": "You are a biomedical QA assistant."},
                {"role": "user", "content": prompt}
            ],
            temperature=0,
            max_tokens=100
        )

        answer = response.choices[0].message.content.strip()
        predictions[item["id"]] = answer
        print(f"[{idx+1}/10] {item['id']} -> {answer}")
        time.sleep(0.5)

    except Exception as e:
        print(f"Error on item {item['id']}: {str(e)}")
        predictions[item["id"]] = ""

# 保存输出
with open("gpt3.5_predict_predictions.json", "w", encoding="utf-8") as f:
    json.dump(predictions, f, indent=2, ensure_ascii=False)
```
