```python
import openai
import json
import time


client = openai.OpenAI(api_key="Your_key")


with open("test.json", "r", encoding="utf-8") as f:
    data = [json.loads(line) for line in f]


predictions = {}


sample_data = data
predictions = {}


model_id = "ft:gpt-3.5-turbo-xxxxxxxxx"

for idx, item in enumerate(sample_data):
    prompt = f"""Context:
{item['context']}

Question:
{item['question']}

Answer:"""

    try:
        response = client.chat.completions.create(
            model=model_id,
            messages=[
                {"role": "user", "content": prompt}
            ],
            temperature=0,
            max_tokens=100
        )
        answer = response.choices[0].message.content.strip()
        predictions[item["id"]] = answer
        print(f"[{idx+1}/{len(data)}] {item['id']} -> {answer}")
        time.sleep(0.5)

    except Exception as e:
        print(f"Error on item {item['id']}: {str(e)}")
        predictions[item["id"]] = ""


with open("gpt_50percent_data_finetune_predictions.json", "w", encoding="utf-8") as f:
    json.dump(predictions, f, indent=2, ensure_ascii=False)
```
